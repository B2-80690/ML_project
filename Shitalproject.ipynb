## Data Pre-Processing

```python
# Function to convert strings to numerical values
def convert_to_numeric(s):
    if isinstance(s, str):
        if s[-1] == 'M':
            return float(s[:-1]) * 1e6
        elif s[-1] == 'K':
            return float(s[:-1]) * 1e3
        else:
            return float(s)
    else:
        return s

# Apply the function to the DataFrame
df['World_Wild_Views'] = df['World_Wild_Views'].apply(convert_to_numeric)
```

```python
# # Remove all square brackets from elements in the 'genre' column
df['List_of_genre'] = df['List_of_genre'].apply(lambda x: ', '.join(x) if isinstance(x, list) else x)
```

```python
df.tail()
```

```python
# Fill missing numeric data with the mean
numeric_columns = ['Run_time_in_min', 'Global_collection', 'Rating_out_of_ten', 'World_Wild_Views', 'Reviews_to_movie']
for col in numeric_columns:
    df[col] = df[col].fillna(df[col].mean())

# Fill missing textual data with the mode
text_columns = ['Name_of_movie','Name_of_director', 'List_of_genre', 'Discription_of_movie', 'Language_of_move', 'Country_of_origin', 'Budget_of_movie']
for col in text_columns:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
# Dummy URL
missing_url = 'https://www.imdb.com/title/tt1192628/mediavie'

# Filling NaN values with the dummy URL
df['Poster_url'] = df['Poster_url'].fillna(missing_url)
```

```python
import re
# Removing textual data from the 'Release_Year' column
# Custom function to remove textual data and replace it with NaN
def clean_release_year(year_string):
    if isinstance(year_string, str):
        pattern = re.compile(r'\d{4}')
        match = pattern.search(year_string)
        if match:
            return match.group()
    return np.nan

# Applying the custom function to the 'Release_Year' column
df['Release_year'] = df['Release_year'].apply(clean_release_year)
```

```python
df['Release_year'] = df['Release_year'].fillna(df['Release_year'].mode()[0])
```

```python
df.isna().sum()
```

```python
# Removing square brackets from the 'List_of_genre' column
df['List_of_genre'] = df['List_of_genre'].apply(lambda x: x.replace("[", "").replace("]", "").replace(", ", "").replace("'", ""))
```
